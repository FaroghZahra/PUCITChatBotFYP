# Stochastic Gradient Descent
Stochastic Gradient Descent is a generic optimization algorithm capable of finding optimal solutions to a wide range of problems. 
The general idea is to tweak parameters iteratively in order to minimize the cost function.
<br>The word `stochastic` means a system or process linked with a random probability. Hence, in Stochastic Gradient Descent, a few samples are selected randomly instead of the whole data set for each iteration.

Stochastic Gradient Descent is when we randomly select a data points on each iteration(forward and backward propagation) when performing the updates to the parameters.

## Code for dummy data

```python
#importing data
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
```

```python
#reading data
data = pd.read_csv('Book1.csv')
data.head()
```

```python
#separating training data and output lables
X_train = data.values[:,0:3]
Y_train = data.values[:,3]
```

```python
w1 = w2 = w3 = w4 = 1
L = 0.01  # The learning Rate
epochs = 1  # The number of iterations to perform gradient descent
n=4
# creating one array for each column
x_train0 = X_train[:,0]
x_train1 = X_train[:,1]
x_train2 = X_train[:,2]
rand = 0  
```

```python
for i in range(epochs): 
    Y_pred = w1 + (w2*(x_train0[rand])) + (w3*(x_train1[rand]) + (w4*(x_train2[rand])))
    D_w1 = (1/n) * (Y_train[rand] - Y_pred)
    D_w2 = (1/n) * x_train0[rand] * (Y_train[rand] - Y_pred)
    D_w3 = (1/n) * x_train1[rand] * (Y_train[rand] - Y_pred)
    D_w4 = (1/n) * x_train2[rand] * (Y_train[rand] - Y_pred)
    w1 = w1 - L * D_w1 
    w2 = w2 - L * D_w2 
    w3 = w3 - L * D_w3 
    w4 = w4 - L * D_w4                 
                       
print(w1)
print(w2)
print(w3)
print(w4)
```
