# LSTM

**Vanishing Gradient Problem**<br>
When number of hidden layers grow, gradient becomes very small and weights will hardly change. When there are more layers in the network, the value of the product of derivative decreases until at some point the partial derivative of the loss function approaches a value close to zero, and the partial derivative vanishes. This will hamper the learning rate.<br>
**Why LSTM?**<br>
We solve Vanishing gradient problem with LSTM. <br>
`LSTM stands for Long-Short Term Memory is also a steping stone to learning about Transformers`<br>
**How can we use a feedback loop to unroll a network that works well with different amount of sequential data?**<br>
In LSTM we can use two diffrent paths instead of one feedback loop to make predictions.One is for **long term memories** and one is for **short term memories**.<br>
<br>Each LSTM state contains:
1. Cell state = Long Term Memory
2. Hidden state
2. Forget gate
3. Input gate
4. Output gate


LSTM uses both sigmoid and tanh functions.<br>
**Sigmoid**<br>
It takes any x-axis coordinate and turn it into y-axis coordinate y-axis coordinate between 0 and 1.<br>
**Tanh**<br>
It takes any x-axis coordinate and turn it into y-axis coordinate y-axis coordinate between -1 and 1.<br>

**<br>How LSTM works?**<br>
1. This is the **cell state** that represents the Long-Term Memory. It can be changed by multiplicaion and then addition. It can't be modified directly beacuse there are no weightd and biases.
This lack of weights alows th LTM to flow through a series of unrolled units without causing the gradient to explode and vanish.
2. The **hidden state** represents the Short-Term Memory. It is directly connected to weights that can modify it.
3. The **forget gate**  determines what percentage of Long-Term Memory will be remembered.
4. The **input gate** combines short term memory and input to create a potential long term memory. and 1st gate determines what percentage of that potential memory to add to the LTM.
5. The **output gate** updates the short term memory. We use new LTM as input  to Tanh function and the return value represents potential short term memory. Now LSTM has to decide how much of this potentiak memory to pass on and this is done using the exact same methd we used earlier.
